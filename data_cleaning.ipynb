```python
data[data["Assets and liabilities"] == "Total assets"]
```

```python
plt.figure(figsize=(16,5))
sns.lineplot(data = data[data["Assets and liabilities"] == "Total assets"], x="REF_DATE", y="VALUE")
plt.xticks(first_month)
plt.xticks(rotation=70)
plt.xlabel("")
plt.ylabel("Assets (in million CAD)")
plt.title("Total Assets (since 2000)")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
original_data = pd.read_csv("data/cleaned_data.csv")
original_data.head()
```

```python
data = original_data.copy()
```

```python
first_month = [
    "2000-01",
    "2001-01",
    "2002-01",
    "2003-01",
    "2004-01",
    "2005-01",
    "2006-01",
    "2007-01",
    "2008-01",
    "2009-01",
    "2010-01",
    "2011-01",
    "2012-01",
    "2013-01",
    "2014-01",
    "2015-01",
    "2016-01",
    "2017-01",
    "2018-01",
    "2019-01",
    "2020-01",
    "2021-01",
    "2022-01",
    "2023-01",
    "2024-01",
]
```

```python
def plot_data(type):
    plt.figure(figsize=(16, 5))
    sns.lineplot(
        data=data[data["Assets and liabilities"] == type],
        x="REF_DATE",
        y="VALUE",
    )
    plt.xticks(first_month)
    plt.xticks(rotation=70)
    plt.xlabel("")
    plt.ylabel("(in million CAD)")
    plt.title(str(type + " (since 2000)"))
    plt.show()
```

```python
plot_data("Total assets")
```

```python
plot_data("Total liabilities")
```

```python
plt.figure(figsize=(16, 5))
sns.lineplot(
    data=data[
        (data["Assets and liabilities"] == "Total assets")
        & (data["REF_DATE"] >= "2020-01")
        & (data["REF_DATE"] <= "2022-12")
    ],
    x="REF_DATE",
    y="VALUE",
)
plt.xticks(rotation=70)
plt.xlabel("")
plt.ylabel("(in million CAD)")
plt.title("Total Assets (COVID years)")
plt.show()
```

```python
plt.figure(figsize=(16, 5))
sns.lineplot(
    data=data[
        (data["Assets and liabilities"] == "Total assets")
        & (data["REF_DATE"] >= "2023-01")
    ],
    x="REF_DATE",
    y="VALUE",
)
plt.xticks(rotation=70)
plt.xlabel("")
plt.ylabel("(in million CAD)")
plt.title("Total Assets (from 2023)")
plt.show()
```

```python
plt.figure(figsize=(16, 5))
sns.lineplot(
    data=data[
        (data["Assets and liabilities"] == "Equity")
        & (data["REF_DATE"] >= "2023-01")
    ],
    x="REF_DATE",
    y="VALUE",
)
plt.xticks(rotation=70)
plt.xlabel("")
plt.ylabel("(in million CAD)")
plt.title("Equity (since 2023)")
plt.show()
```

```python
plot_data("Total, Government of Canada, direct and guaranteed securities")
```

```python
plot_data("Government of Canada, Treasury Bills")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("data/cleaned_data.csv")
```

```python
first_month = [
    "2000-01",
    "2001-01",
    "2002-01",
    "2003-01",
    "2004-01",
    "2005-01",
    "2006-01",
    "2007-01",
    "2008-01",
    "2009-01",
    "2010-01",
    "2011-01",
    "2012-01",
    "2013-01",
    "2014-01",
    "2015-01",
    "2016-01",
    "2017-01",
    "2018-01",
    "2019-01",
    "2020-01",
    "2021-01",
    "2022-01",
    "2023-01",
    "2024-01",
]
```

```python
def plot_data(type):
    plt.figure(figsize=(10, 5))
    sns.lineplot(
        data=data[data["Assets and liabilities"] == type],
        x="REF_DATE",
        y="VALUE",
    )
    plt.xticks(first_month)
    plt.xticks(rotation=70)
    plt.xlabel("")
    plt.ylabel("(in million CAD)")
    plt.title(str(type + " (since 2000)"))
    plt.show()


def plot_data_COVID(type):
    plt.figure(figsize=(10, 5))
    sns.lineplot(
        data=data[
            (data["Assets and liabilities"] == type)
            & (data["REF_DATE"] >= "2020-01")
            & (data["REF_DATE"] <= "2022-12")
        ],
        x="REF_DATE",
        y="VALUE",
    )
    plt.xticks(rotation=70)
    plt.xlabel("")
    plt.ylabel("(in million CAD)")
    plt.title(str(type + " (COVID years)"))
    plt.show()


def plot_data_2023(type):
    plt.figure(figsize=(10, 5))
    sns.lineplot(
        data=data[
            (data["Assets and liabilities"] == type) & (data["REF_DATE"] >= "2023-01")
        ],
        x="REF_DATE",
        y="VALUE",
    )
    plt.xticks(rotation=70)
    plt.xlabel("")
    plt.ylabel("(in million CAD)")
    plt.title(str(type + " (since 2023)"))
    plt.show()
```

```python
plot_data("Total assets")
```

```python
plot_data_COVID("Total assets")
```

```python
plot_data_2023("Total assets")
```

Issued with terms to maturity between 2 and 30 years, government bonds are considered very low-risk fixed income investments as they are backed by governments. The value of government bonds fluctuates based on supply and demand in the market  a government will increase the supply of bonds to raise money, which will be used to stimulate the economy. Demand for government bonds tends to increase during periods of low confidence in equity markets as investors seek safety. Demand also tends to increase in periods of weak economic activity when the threat of inflation is minimized.

Treasury bills (T-Bills) and other money market securities such as commercial paper and bankers acceptances are considered to be the safest segment of the fixed income market. Issued at a discount, T-Bills are short-term debt securities issued or guaranteed by federal, provincial or other governments. The stated interest rates for T-Bills are fixed when issued, but values fluctuate based on changes to the central bank rate. A T-Bills return is calculated based on the difference between the price paid and the par value (also known as the denomination or face value). T-Bills mature at par (typically 90 or 180 days) and do not pay fixed interest payments like most bonds.

During the COVID-19 pandemic, **total assets and liabilities** of Canadian households, businesses, and the government rose significantly due to several factors, and their reduction since 2023 can be explained by the reversal of these pandemic-driven trends. Let's break down both periods:

### Why Total Assets and Liabilities Rose During COVID-19:

1. **Government Stimulus and Debt Programs**:
   - The Canadian government introduced **massive fiscal stimulus** packages to support businesses, households, and healthcare systems. Programs like the **Canada Emergency Response Benefit (CERB)** and wage subsidies injected significant liquidity into the economy.
   - To finance these measures, the government issued a large amount of debt, increasing its **liabilities**. This also led to an increase in **public sector assets** as it acquired resources and injected funds into the economy.

2. **Monetary Policy and Low Interest Rates**:
   - The Bank of Canada cut interest rates to historically low levels to stimulate borrowing and spending. This led to an increase in **household and corporate borrowing** (liabilities) as credit was cheap.
   - Low rates also drove up asset prices, particularly in **real estate and financial markets**, as investors sought higher returns in riskier assets, raising the value of total assets.

3. **Rising Housing Prices**:
   - During the pandemic, **housing prices surged** due to high demand, low interest rates, and a desire for more space during lockdowns. This caused household assets to rise, as real estate is a significant component of Canadians' total assets.
   - At the same time, many homeowners took on **larger mortgages**, further increasing liabilities.

4. **Business Borrowing**:
   - Many businesses took on **additional debt** to survive the economic shutdowns, either through direct borrowing or government-backed loan programs, driving up corporate liabilities.

### Why Assets and Liabilities Are Reducing Since 2023:

1. **Rising Interest Rates**:
   - To combat inflation, which spiked after the pandemic due to supply chain disruptions and strong demand, the Bank of Canada started **raising interest rates** significantly in 2022 and 2023. Higher interest rates make borrowing more expensive, slowing down new debt accumulation (liabilities) for both businesses and households.
   - Rising rates also contributed to a **cooling of the housing market**, leading to a reduction in housing prices, which in turn reduces household assets.

2. **Debt Repayment**:
   - As the economy has recovered, businesses and households may be focusing on **deleveraging** or paying off their pandemic-era debts, reducing total liabilities.

3. **Asset Price Corrections**:
   - The post-pandemic period has seen **corrections in asset prices**, particularly in real estate and financial markets. As housing prices stabilize or even decline and stock market gains are moderated, the **total value of assets** declines.
   
4. **Winding Down of Government Programs**:
   - Many of the emergency fiscal and monetary programs implemented during the pandemic have been phased out. This means the government is borrowing less, which reduces its liabilities.
   - Additionally, the unwinding of asset purchases by the Bank of Canada (such as selling off government bonds) reduces the central banks balance sheet.

5. **Inflation Control Measures**:
   - As inflationary pressures are being addressed, economic growth has slowed, which impacts household wealth accumulation and business expansion, leading to more stable or declining asset values.

### Summary:
- **During COVID-19**: Total assets rose due to increased asset values (especially housing and stocks) and government interventions, while liabilities increased as both the public and private sectors took on more debt.
- **Since 2023**: Rising interest rates, a cooling real estate market, debt repayment, and the end of government stimulus measures have led to a reduction in total assets and liabilities.

This reflects a broader global trend of economies adjusting after the unprecedented financial and policy interventions of the pandemic.

The Canadian government launched **Treasury bills (T-bills)** during the COVID-19 pandemic as a response to the economic crisis caused by the pandemic, and then stopped issuing new ones after 2021 for several key reasons tied to both the immediate financial needs of the pandemic and the post-pandemic recovery. Let's break this down:

### Why the Canadian Government Launched T-Bills During COVID:

1. **Raising Emergency Funds Quickly**:
   - T-bills are short-term debt instruments that governments issue to quickly raise cash. The Canadian government needed **immediate liquidity** to finance emergency spending programs during the pandemic, such as the **Canada Emergency Response Benefit (CERB)**, wage subsidies, and healthcare support. 
   - T-bills allowed the government to **borrow large sums of money** in the short term to fund these programs while maintaining flexibility for future borrowing or repayment.

2. **Lower Interest Costs**:
   - T-bills typically have very short maturities (e.g., 3 months, 6 months, or 1 year), and during the pandemic, interest rates were at historic lows due to the Bank of Canada's aggressive monetary policy.
   - The government took advantage of these **ultra-low interest rates** to issue T-bills at **minimal borrowing costs**, making it a cheap and effective way to raise funds quickly.

3. **Managing Cash Flow**:
   - The government's cash needs during the pandemic were **highly uncertain**, as the length and severity of the crisis were difficult to predict. T-bills provided a flexible tool for **short-term cash management**, allowing the government to adjust its borrowing depending on how the situation evolved.

4. **Increased Investor Demand**:
   - During periods of economic uncertainty, like the COVID-19 pandemic, investors tend to seek out **safe, liquid assets**, such as government T-bills. This increased demand for T-bills allowed the government to issue more of them without significantly raising interest rates or borrowing costs.

### Why the Canadian Government Stopped Issuing New T-Bills After 2021:

1. **Economic Recovery and Lower Borrowing Needs**:
   - By the end of 2021, the Canadian economy was **starting to recover** from the pandemic. Economic activity resumed, and the immediate need for emergency government spending began to decline.
   - As the **fiscal deficit shrank** and emergency programs were wound down, the governments need for short-term financing diminished, reducing the necessity for issuing T-bills.

2. **Shift to Longer-Term Borrowing**:
   - As the economy stabilized, the government shifted from short-term borrowing (T-bills) to **longer-term bonds**. Longer-term bonds allow the government to lock in **favorable interest rates** for extended periods, reducing refinancing risk and smoothing out future debt repayments over many years.
   - Issuing longer-term debt also aligns with a more **stable, post-crisis fiscal environment**, where short-term borrowing is no longer as critical.

3. **Rising Interest Rates**:
   - By 2022 and 2023, the Bank of Canada began **raising interest rates** to combat inflation, making borrowing more expensive. T-bills, being short-term instruments, are more **sensitive to rising interest rates**. As rates increased, the government likely found it more advantageous to move away from short-term T-bills toward **longer-term debt instruments**, which could lock in rates before they rose further.

4. **Debt Management Strategy**:
   - T-bills are useful for addressing short-term cash needs but require frequent refinancing, which can be risky if interest rates rise or market conditions change quickly. After 2021, the governments **debt management strategy** likely focused on reducing its reliance on short-term instruments in favor of more predictable and stable financing structures.

### Summary:
- **During COVID-19**: The Canadian government launched T-bills to meet urgent and substantial cash flow needs, taking advantage of low interest rates and strong demand for safe assets during the economic crisis.
- **After 2021**: As the economy recovered, emergency programs ended, and interest rates began rising, the government shifted its focus to longer-term borrowing to manage debt more sustainably and reduce refinancing risks.

This transition reflects a typical move from short-term crisis management to long-term financial stability.
